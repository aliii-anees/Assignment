```python
import sklearn
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv('dataset_main.csv')
```

```python
datasetmod = dataset.copy(deep = True)
datasetmod[['Age','Gender','sudden weight loss','weakness','class']] = datasetmod[['Age','Gender','sudden weight loss','weakness','class']].replace(0,np.NaN)
```

```python
datasetmod.isnull().sum()
```

```python
datasetmod['Gender'].replace(['Male', 'Female'],
                        [1, 0], inplace=True)
datasetmod['sudden weight loss'].replace(['Yes', 'No'],
                        [1, 0], inplace=True)
datasetmod['weakness'].replace(['Yes', 'No'],
                        [1, 0], inplace=True)
datasetmod['Obesity'].replace(['Yes', 'No'],
                        [1, 0], inplace=True)
datasetmod['class'].replace(['Positive', 'Negative'],
                        [1, 0], inplace=True)
```

```python
dataset.head(7)
```

```python
p = dataset.hist(figsize = (20, 20))
```

```python
dataset['Age'].fillna(dataset['Age'].mean(), inplace = True)
dataset['Gender'].fillna(dataset['Gender'].mean(), inplace = True)
dataset['sudden weight loss'].fillna(dataset['sudden weight loss'].median(), inplace = True)
dataset['weakness'].fillna(dataset['weakness'].median(), inplace = True)
dataset['Obesity'].fillna(dataset['Obesity'].median(), inplace = True)
dataset['class'].fillna(dataset['class'].median(), inplace = True)
```

```python
Q1 = dataset['Age'].quantile(0.25)
Q3 = dataset['Age'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR
print(IQR)
print (lower)
print (upper)
```

```python
upper_array = np.where(dataset['Age']>=upper)[0]
lower_array = np.where(dataset['Age']<=lower)[0]
```

```python
dataset.drop(index=upper_array, inplace=True)
dataset.drop(index=lower_array, inplace=True)
```

```python
plt.figure(figsize=(12,10))  
p=sns.heatmap(dataset.corr(), annot=True,cmap ='RdYlGn')
```

```python
XSC = StandardScaler()
X = pd.DataFrame(XSC.fit_transform(dataset.drop(["class"], axis = 1), ), 
              columns = ['Age', 'Gender', 'sudden weight loss', 'weakness', 'Obesity'])
```

```python
X.head()
```

```python
y = dataset['class']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=1/3,random_state=42, stratify=y)
```

```python
test_scores = []
train_scores = []

for i in range(1,29):

    knn = KNeighborsClassifier(i)
    knn.fit(X_train,y_train)
    
    train_scores.append(knn.score(X_train,y_train))
    test_scores.append(knn.score(X_test,y_test))
```

```python
max_train_score = max(train_scores)
train_scores_ind = [i for i, v in enumerate(train_scores) if v == max_train_score]
print('Max train score {} % and k = {}'.format(max_train_score*100,list(map(lambda x: x+1, train_scores_ind))))
```

```python
max_test_score = max(test_scores)
test_scores_ind = [i for i, v in enumerate(test_scores) if v == max_test_score]
print('Max test score {} % and k = {}'.format(max_test_score*100,list(map(lambda x: x+1, test_scores_ind))))
```

```python
plt.figure(figsize=(12, 5))

p = sns.lineplot(train_scores, marker='*', label='Train Score')
p = sns.lineplot(test_scores,marker='o',label='Test Score')
```

```python
knn = KNeighborsClassifier(1)

knn.fit(X_train,y_train)
knn.score(X_test,y_test)
```
